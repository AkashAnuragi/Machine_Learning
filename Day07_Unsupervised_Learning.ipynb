# Unsupervised Learning (Unlabeled Dataset (No Output/No Label))
Unsupervised learning is type of ML which is not like supervised with proper data points ans label this ML is without label and unappropriate data point, we use this data for finding patterns, relationship and mostly the groups
                                                                                                                                                                               
                                                                                                                                                                                               
### Types of Unsupervised Learning
  Clustering
    Grouping  
- Dimension Reduction
  - Reduce the number of features/columns
- Association Rules
   - Associativity/Relationship in data points
 - Anomaly Detection
   - Fraud Detection, Unusal activity
  
### Clustering in Unsupervised Learning
#### Grouping similar types of data point without label on the basis of feature's characterstics 
1- K-Means Clustering
   

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = sns.load_dataset('iris')
```

```python
df.shape
```

```python
df.head(2)
```

```python
plt.figure(figsize = (12,3))
sns.scatterplot(x = 'petal_length', y= 'petal_width', data =df)
plt.show()
```

```python
plt.figure(figsize = (6,3))
sns.scatterplot(x = 'petal_length', y= 'petal_width',  hue = 'species', data =df)
plt.show()
```

```python
X = df.drop(columns = ['species'])
```

```python
X.head(2)
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(
    n_clusters = 3,
    random_state = 42
)
```

```python
kmeans.fit(X)
```

```python
label = kmeans.labels_
```

```python
label
```

```python
plt.figure(figsize=(6,3))
sns.scatterplot(x='petal_length', y='petal_width', hue = label , data=df)
plt.show()
```

```python
plt.figure(figsize =(12,3))
plt.subplot(1,2,1)
sns.scatterplot(x = 'petal_length', y= 'petal_width', hue = 'species' ,data = df)
plt.subplot(1,2,2)
sns.scatterplot(x='petal_length', y='petal_width', hue = label , data=df)
plt.show()
```

```python
kmeans.cluster_centers_
```

```python
plt.figure(figsize=(12,3))
plt.subplot(1,2,1)
plt.scatter(X['petal_length'], X['petal_width'] )
plt.subplot(1,2,2)
plt.scatter(X['petal_length'], X['petal_width'], c= label)
plt.scatter(kmeans.cluster_centers_[:,0],  kmeans.cluster_centers_[:,1], marker = 'X' ,s =200)
plt.show()
```
