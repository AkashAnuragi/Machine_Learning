```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = df.bfill()
```

```python
df.isnull().sum().sum()
```

```python
df['Sale Price'].unique()
```

```python
df[['Sale Price','Gross Sales']].describe()
```

```python
# Standardization
# mean =0, sts =1
# z = (x-miu(mean))/sigma(std)
```

```python
li = ['Sale Price', 'Gross Sales']

for col in li:
    df[col] = (df[col]-df[col].mean())/df[col].std()
```

```python
df[['Sale Price','Gross Sales']].describe()
```

```python
df = pd.read_excel(r"F:\Machine Learning\Machine_Learning\DataModel\Financial_Sample.xlsx")
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
```

```python
output = pd.DataFrame(sc.fit_transform(df[['Sale Price','Gross Sales']]),columns=['Sale Price','Gross Sales'])
```

```python
output.describe()
```

```python
# Normalization
# Convert all Values  between 0 to 1
# n = (X - Xmin)/(Xmax-Xmin)
```

```python
li = ['Sale Price','Gross Sales']
for col in li:
    df[col] = (df[col]-df[col].min())/(df[col].max() - df[col].min())
```

```python
df[['Sale Price','Gross Sales']].describe()
```
