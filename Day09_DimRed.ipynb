```python
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
df = sns.load_dataset('iris').ffill()
```

```python
df.head(2)
```

```python
X = df.drop(columns = ['species'])
```

```python
X.head(2)
```

### Dimensional Reduction
1. Feature Selection
   - Remove unwanted/useless columns (keeps only valuesable columns)
3. Feature Extraction
   - Combines two or more than two columns without loosing its properties
#### After Dimensional Reduction
- Reduce Computational Time

- Visualize High-dimensional graphs
- Improve the model's Performance
### PCA (Principle Component Analyis)
- Capture High/maximum Variance
- Are perpendicular to each other

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
X_scaled = scaler.fit_transform(X)
```

```python
X_scaled[:4]
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2)
```

```python
X_pca = pca.fit_transform(X_scaled)
```

```python
X_pca[:4]
```

```python
pca.explained_variance_ratio_
```

```python
sum(pca.explained_variance_ratio_)
```

```python
plt.figure(figsize=(6,3))
plt.scatter( X_pca[:,0],X_pca[:,1] )
plt.show()
```
