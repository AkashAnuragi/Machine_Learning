### ML  (Machine Learning)
- Machine Learning is a technique by  a computer Learns pattern from the historical/past dataset, so that it can predict a next value.

- Machine Learning is a mathematical Formula.

<b> Traditional Programming</b><br>
Program + input => Output </br>
<b>Machine Learning Programming</b></br>
Input + Output => Program then, </br>
Program(ML) + Input(New) => Output(Predicted)</br>

<b> Example </b> </br>
Input - Output</br>
0 - 0 => 0 </br>
2 - 2 => 4 </br>

#### Types of Machine Learning
1. Supervised Learning ( labeled dataset (Input + Output))
    - Regression     (Predict Contiguous value (House Price Prediction , Stock Price Prediction etc))
    - Classification (Classify/ctaegorize a group Yes/No , True/False , 0/1 , Spam/Not Spam etc)
2. Unsupervised Learning (Unlabeled dataset (Input))
    - Dimensional Reduction (Reduce the number of feature)
        - Feature Selection ( Remove the unwanted column , useless feature )
        - Feature Extraction ( Merge Feature : Algo  PCA , t-SNE , UMAP , LDA etc )
    - Clustering (Grouping Algo : KMeans , DBSCAN etc)
    - Association Rules ( Relationship b/w two datapoints , itemsets (Recommendation Model) )
3. Reinforcement Learning (Without dataset)
    - Trial n Error ( Reward and Punishment )
    - video games , Autopilot etc
